```python
import pandas as pd #
df = pd.read_csv('sample_data/samples.csv')
df.head()
df_top10 = df.loc[ (df['winner_rank'] <= 10) ]
df_top10
a = df_top10.groupby('winner_name').mean()
a
# Stan Wawrinka, 'Jack Sock' 'David Goffin', 'Dominic Thiem', 제외해야할 인물들
df_top10_mean = a.drop(['Stan Wawrinka', 'Jack Sock', 'David Goffin', 'Dominic Thiem']) # drop ( row list (index or row name))
df_top10_mean

df_HC_all = df.loc[ df['winner_name'] == 'Hyeon Chung']
df_HC_mean = df_HC_all.describe().loc[['mean']]
df_HC_mean
## Load All Data
## Filter by Top 10 Ranker
## Grouped by 'winner_name'
##  

# df[['latitude']] # 컬럼 전체 필요할때, 
# a = df.loc[df['total_rooms'] > 10000, ['total_rooms', 'total_bedrooms']] # 컬럼에서 Row Filtering 하고 뽑고 싶을때
# b = a.iloc[:5]
# b
# a.iloc[:5].describe().loc[['mean']] ## 선수 한명 평균데이터 뽑을때

## Make Top 10 Average stat
## Make chung Hyun Stat
## Make Each Player Stat by overlapping
```

```python
# 1. 첫 서브 성공시 득점율
df_top10_filter = pd.DataFrame(df_top10_mean['w_1stWon'] / df_top10_mean['w_1stIn'], columns=["1st_serv_point"])
df_top10_filter
df_HC_filter = pd.DataFrame(df_HC_mean['w_1stWon'] / df_HC_mean['w_1stIn'], columns=["1st_serv_point"])
df_HC_filter
```

```python
# 2. 첫 리턴 득점율
df_top10_filter['1st_ret_point'] = (df_top10_mean['l_1stIn'] - df_top10_mean['l_1stWon']) / df_top10_mean['l_1stIn']
df_top10_filter
df_HC_filter['1st_ret_point'] = (df_HC_mean['l_1stIn'] - df_HC_mean['l_1stWon']) / df_HC_mean['l_1stIn']
df_HC_filter
```

```python
# 에이스 갯수
df_top10_filter['ace_ratio'] = df_top10_mean['w_ace'] / df_top10_mean['w_svpt']
df_top10_filter
df_HC_filter['ace_ratio'] = df_HC_mean['w_ace'] / df_HC_mean['w_svpt']
df_HC_filter
```

```python
# 키(height) // 200분의 1로 단위 조정
df_top10_filter['winner_ht'] = df_top10_mean['winner_ht'] / 200
df_top10_filter
df_HC_filter['winner_ht'] = df_HC_mean['winner_ht'] / 200
df_HC_filter
```

```python
# 5. 브레이킹 성공률
df_top10_filter['breaking_won'] = df_top10_mean['w_bpSaved'] / df_top10_mean['w_bpFaced']
df_top10_filter
df_HC_filter['breaking_won'] = df_HC_mean['w_bpSaved'] / df_HC_mean['w_bpFaced']
print(df_HC_filter)

df_top10_avg = df_top10_filter.describe().loc[['mean']]
print(df_top10_avg)
```

# 새 섹션

```python
# 방사형 그래프 그리기
import numpy as np
import matplotlib.pyplot as plt

categories = list(df_top10_avg.columns)
categories

from math import pi
num_labels = len(categories)
my_palette = plt.cm.get_cmap("Set2", 3) ## Make Each Categories Color map 
angles = [x/float(num_labels)*(2*pi) for x in range(num_labels)] ## 각 등분점
# print(angles)
fig = plt.figure(figsize=(12,12))
dfs = [df_top10_avg, df_HC_filter]
names = ['Average Top 10 Players', 'Hyeon Chung']
for i,v in enumerate(dfs):
  color = my_palette(i)
  data = v.iloc[0].tolist()
  # print("normliazed data: ", data)
  ax = plt.subplot(polar=True)
  plt.xticks(angles, categories, fontsize=13)
  ax.tick_params(axis='x', which='major', pad=15) ## x축과 눈금 사이에 여백을 준다.
  
  # plt.ylim(0,8) # y 축 눈금 간격 조절
  ax.set_theta_offset(pi / 2)
  ax.set_theta_direction(-1)
  ax.plot(angles, data, color=color, linewidth=2, label=names[i], linestyle='solid')
  ax.fill(angles, data, color=color, alpha=0.4)
plt.title('Average Top 10 Player vs Hyeon Chung')
plt.legend()
plt.show()
```
