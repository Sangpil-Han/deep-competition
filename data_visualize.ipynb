```python
# df.loc[:9][['w_1stWon','w_1stIn','l_1stIn','w_ace']]
# df = pd.read_csv('data/atp_matches_2018.csv')
## Rank 10 위 까지 필터링
top10stats = df.sort_values(by='winner_rank')
top10df=top10stats.loc[top10stats['winner_rank']<= 10]
# top10df['winner_name']
top10df
```

```python
# 1. 첫서브 성공시 득점율
df_top10 = pd.DataFrame(top10df['w_1stWon'] / top10df['w_1stIn'], columns=["1st_serv_point"])
df_top10['winner_name'] = top10df['winner_name']
df_top10['winner_rank'] = top10df['winner_rank']
df_top10
```

```python
# 첫 리턴 득점율
df_top10['1st_ret_point'] = (top10df['l_1stIn'] - top10df['l_1stWon']) / top10df['l_1stIn']
df_top10
```

```python
# 에이스 갯수
df_top10['ace_ratio'] = top10df['w_ace'] / top10df['w_svpt']
# df_top10['w_ace'] = top10df['w_ace']
#df_top10.drop(['ace_ratio'])
#df_top10 = df_top10.drop(columns=['w_ace'])
df_top10
```

```python
# 키(height)
top100stats = df.sort_values(by='winner_rank')
top100df=top100stats.loc[top100stats['winner_rank']<= 100]
# top100df
top100df['winner_ht'].mean() # 100명의 키에 대한 평균값
df_top10['winner_ht'] = top10df['winner_ht']
df_top10
```

```python
# 브레이킹 성공률
df_top10['breaking_won'] = top10df['w_bpSaved'] / top10df['w_bpFaced']
df_top10
# print(df_top10)
# print(df_top10['w_ace'].mean())
# Data Normalization
# normalized_df=(df_top10-df_top10.mean())/df_top10.std()+5
# Data NaN 어떻게 채워줄지?
```
